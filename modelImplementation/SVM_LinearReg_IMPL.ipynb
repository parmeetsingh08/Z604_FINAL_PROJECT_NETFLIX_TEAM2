```python
from sklearn.svm import SVC
```

```python
def mySVMModel(feature_matrix_train, y_train, feature_matrix_test):
    clf = SVC()
    clf.fit(feature_matrix_train, y_train)  #.set_params(kernel='linear')
    return clf.predict(feature_matrix_test)
```

```python
from sklearn import linear_model
def myLinearModel(feature_matrix_train, y_train, feature_matrix_test):
    logreg = linear_model.LogisticRegression(C=1e5)
    logreg.fit(feature_matrix_train, y_train)
    return logreg.predict(feature_matrix_test)
```

```python

def get_word2vec_feature_for_movie(mv_id):
    return df_movie[df_movie.MovieID==mv_id]
    
def myModel(group):
    op = []
    result = []
    for row in group.itertuples():
        var = get_word2vec_feature_for_movie (row.MovieID)
        op.append (var)
    
    op = pd.concat(op)
    try:
        ##Comment/Uncomment as per the required model implementation
        #return mySVMModel(op.iloc[:,1:201], group['user_rating'],df_movie.iloc[:,1:201])
        return myLinearModel(op.iloc[:,1:201], group['user_rating'],df_movie.iloc[:,1:201])
    except ValueError :
        print "error------"
        print group
    #return op
        
#temp= df_user_rating.groupby('UserID').apply(mySVM)
temp= df_user_rating_sample.groupby('UserID').apply(myModel)
```
