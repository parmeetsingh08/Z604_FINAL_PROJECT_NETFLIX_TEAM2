```python
def getRecommendations(group):
    recommendations_movies = []
    #print "###########"
    workingset = group.sort('user_rating', ascending=False).head(3)
    #print workingset
    for row in workingset.itertuples():
        #intermediate = neighbors_with_distane[neighbors_with_distane.MovieID==row.MovieID]
        #print intermediate
        #print str(intermediate.NeighborID)
        recommendations_movies.append(neighbors_with_distane[['NeighborID','Distance']][neighbors_with_distane.MovieID==row.MovieID])
    recommendations_movies = pd.concat(recommendations_movies,ignore_index=True)
    return recommendations_movies
```

```python
temp_new = df_user_rating_sample.groupby('UserID').apply(getRecommendations)
#temp_new
```

```python
final_df = temp_new.reset_index(0).drop_duplicates().sort(['UserID','Distance'],ascending=True)
```
